```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style('white')
import random
```

```python
#use %matplotlib for graph to appear in another screen
%matplotlib inline

#create figure and axes objects
fig,ax = plt.subplots(figsize = (7,7))


#formatting the grid appearance
ax.grid(True);
ax.set_xlim(0,95);
ax.set_ylim(0,95);
ax.set_xticks(np.arange(0,100,5));
ax.set_yticks(np.arange(0,100,5));

#plotting points - code can be made better
y = np.array([60]*6)
x = np.linspace(25,50,6)
ax.plot(x,y,'bo-');

y = np.array([45]*6)
x = np.linspace(25,50,6)
ax.plot(x,y,'bo-');

y = [50,55,45,60]
x = [25]*4
ax.plot(x,y, 'bo-');
```
